```python
import numpy as np
import matplotlib.pyplot as plt

# set up the seed
np.random.seed(6644)

# define the number of samples
n = int(1e5)

# generate Z
Z = np.random.normal(0, 1, n)

# display summary
print("Summary of Z:")
print("Mean: ", np.mean(Z))
print("Standard Deviation: ", np.std(Z))
print("Min: ", np.min(Z))
print("Max: ", np.max(Z))

# generate histogram
plt.hist(Z, bins=30)
plt.title("Histogram of Z")
plt.xlabel("Z")
plt.show()

# initialize Z1_polar and Z2_polar
Z1_polar = np.zeros(n)
Z2_polar = np.zeros(n)

# generate Z1_polar and Z2_polar
for i in range(n):
  while True:
    V1 = 2*np.random.uniform() - 1
    V2 = 2*np.random.uniform() - 1
    W = V1**2 + V2**2
    if W <= 1: break
  Y = np.sqrt((-2 * np.log(W)) / W)
  Z1_polar[i] = V1*Y
  Z2_polar[i] = V2*Y

# display summary of Z1_polar
print("\nSummary of Z1_polar:")
print("Mean: ", np.mean(Z1_polar))
print("Standard Deviation: ", np.std(Z1_polar))
print("Min: ", np.min(Z1_polar))
print("Max: ", np.max(Z1_polar))

# generate histogram of Z1_polar
plt.hist(Z1_polar, bins=30)
plt.title("Histogram of Z1_polar")
plt.xlabel("Z1_polar")
plt.show()

# display summary of Z2_polar
print("\nSummary of Z2_polar:")
print("Mean: ", np.mean(Z2_polar))
print("Standard Deviation: ", np.std(Z2_polar))
print("Min: ", np.min(Z2_polar))
print("Max: ", np.max(Z2_polar))

# generate histogram of Z2_polar
plt.hist(Z2_polar, bins=30)
plt.title("Histogram of Z2_polar")
plt.xlabel("Z2_polar")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pydistribution.distribution import *

n = int(1e5)

def standard_normal(prng=DEFAULT_PRNG, u=None):
    if u is None:
        u = prng()
    u = prn_handler(u)

    def sign(x):
        if x > 0:
            return 1
        if x == 0:
            return 0
        return -1

    t = abs((-math.log(min(u, 1 - u)) ** 2) ** (0.5))

    c0 = 2.515517
    c1 = 0.802853
    c2 = 0.010328
    d1 = 1.432788
    d2 = 0.189269
    d3 = 0.001308

    return sign(u - 0.5) * (
        t
        - 
        (
            (c0 + (c1 * t) + (c2 * (t**2)))
            / (1 + (d1 * t) + (d2 * (t**2)) + d3 * (t**3))
        )
    )



Z = [standard_normal() for i in range(n)]
pairs = [standard_normal_pair() for i in range(n)]
Z1 = [i[0][0] for i in pairs]
Z2 = [i[0][1] for i in pairs]


```
