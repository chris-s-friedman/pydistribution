```python
import matplotlib.pyplot as plt
import numpy as np
from pydistribution.distributions import *



def rv_plotter(rv, rv_name, bins= 100):
    fig, ax = plt.subplots()
    rv_mean = np.mean(rv)
    rv_std = np.std(rv)
    rv_min = np.min(rv)
    rv_max = np.max(rv)
    # generate histogram
    ax.hist(rv, bins = bins)
    fig.suptitle(f"Histogram of {rv_name}")
    ax.set_xlabel(rv_name)
    ax.text(x=0.95, y=0.95, 
        s = f'Mean: {rv_mean:.2f}\nStd: {rv_std:.2f}\nMin: {rv_min:.2f}\nMax: {rv_max:.2f}',
        horizontalalignment='right',
        verticalalignment='top',
        transform=ax.transAxes)
    return plt

def rv_function_plotter(func, generate_n = int(1e5), **kwargs):
    print("-" * 80)
    print(func.__name__)
    for key, value in kwargs.items():
        print(f"Key: {key}, Value: {value}")
    rv = [func(**kwargs) for i in range(generate_n)]
    return rv_plotter(rv, func.__name__, )
```

```python
rv_function_plotter(weibull, lam = 1, b = 2).show()
rv_function_plotter(exponential, lam = 1).show()
rv_function_plotter(laplace, mu = 0, b = 5).show()
rv_function_plotter(triangular).show()
rv_function_plotter(triangular, minimum=-40, mode=10, maximum=60).show()
rv_function_plotter(bernouli, p = .5).show()
rv_function_plotter(bernouli, p = .25).show()
rv_function_plotter(geometric, p = .5).show()
rv_function_plotter(geometric, p = .25).show()
rv_function_plotter(poisson, lam = 1).show()
rv_function_plotter(binomial, n = 500, p = .5).show()
rv_function_plotter(binomial, n = 500, p = .25).show()
rv_function_plotter(erlang, lam = 1,n = 500).show()
rv_function_plotter(negative_binomial, n = 500, p = .5).show()
```
